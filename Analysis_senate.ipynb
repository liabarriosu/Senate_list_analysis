Ahora pasaremos a números las columnas que queramos. En el caso de la columna "REMUNERACION TOTAL", es importante, limpiar antes los datos.
1- Pasar a texto la columna, para poder usar las funciones de texto.
2- Eliminar los espacios en blanco al inicio y al final.
3- Eliminar comas
4- Reemplazar posibles valores vacíos con NaN.

Finalmente, podemos convertir la columna a una columna numérica.

```python
# Limpiar la columna 'REMUNERACION TOTAL' antes de convertir a numérico

# Convertir a string para asegurar que podemos usar métodos de string
df['REMUNERACION TOTAL'] = df['REMUNERACION TOTAL'].astype(str)

# Eliminar espacios en blanco al inicio y final
df['REMUNERACION TOTAL'] = df['REMUNERACION TOTAL'].str.strip()

# Eliminar las comas (si las hay)
df['REMUNERACION TOTAL'] = df['REMUNERACION TOTAL'].str.replace(',', '', regex=False)

# Reemplazar posibles valores vacíos o no numéricos restantes con NaN antes de la conversión
df['REMUNERACION TOTAL'] = df['REMUNERACION TOTAL'].replace('', pd.NA)


# Convertir la columna 'REMUNERACION TOTAL' a numérica.
# Usamos errors='coerce' para convertir cualquier valor que aún no sea numérico a NaN
df['REMUNERACION TOTAL'] = pd.to_numeric(df['REMUNERACION TOTAL'], errors='coerce')


```

Ahora cambiamos las demás columnas que nos interesan a números. Por ejemplo "DEVENGADO", repitiendo el procedimiento.

```python
# Limpiar la columna 'DEVENGADO' antes de convertir a numérico

# Convertir a string para asegurar que podemos usar métodos de string
df['DEVENGADO'] = df['DEVENGADO'].astype(str)

# Eliminar espacios en blanco al inicio y final
df['DEVENGADO'] = df['DEVENGADO'].str.strip()

# Eliminar las comas (si las hay)
df['DEVENGADO'] = df['DEVENGADO'].str.replace(',', '', regex=False)

# Reemplazar posibles valores vacíos o no numéricos restantes con NaN antes de la conversión
df['DEVENGADO'] = df['DEVENGADO'].replace('', pd.NA)


# Convertir la columna 'DEVENGADO' a numérica.
# Usamos errors='coerce' para convertir cualquier valor que aún no sea numérico a NaN
df['DEVENGADO'] = pd.to_numeric(df['DEVENGADO'], errors='coerce')
```
